```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Define nodes
nodes = [
    ("start", {"type": "start", "name": "Start Node"}),
    ("end", {"type": "end", "name": "End Node"}),
    ("tool_1", {"type": "tool", "name": "Tool Node 1"}),
    ("tool_2", {"type": "tool", "name": "Tool Node 2"}),
    ("tool_3", {"type": "tool", "name": "Tool Node 3"}),
    ("llm_1", {"type": "llm", "name": "LLM Node 1"}),
    ("llm_2", {"type": "llm", "name": "LLM Node 2"})
]

# Add nodes to the graph
G.add_nodes_from(nodes)

# Define edges (connections between nodes)
edges = [
    ("start", "tool_1"),
    ("tool_1", "llm_1"),
    ("llm_1", "tool_2"),
    ("tool_2", "llm_2"),
    ("llm_2", "tool_3"),
    ("tool_3", "end")
]

# Add edges to the graph
G.add_edges_from(edges)

# Draw the graph
pos = nx.spring_layout(G)  # Use spring layout for positioning nodes
plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="lightblue", font_size=10, font_weight="bold", arrows=True)
plt.title("Langgraph: Start, Tools, LLMs, and End Nodes")
plt.show()
```
